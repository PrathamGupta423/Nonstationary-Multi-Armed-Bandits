```python
import numpy as np
from tqdm import tqdm
import json
import matplotlib.pyplot as plt
```

# get the JSON

have saved tons of json files, get all of them in order of:
1. static environment
2. global environment
3. abrupt enviornment
4. abrupt local environment.

```python
# let's get the json file for static first:
static_paths = [ 'JSON/static_ADR_KLUCB_results.json','JSON/static_ADR_TS_results.json',
                'JSON/static_ADS_TS_results.json',
                'JSON/static_Discounted_UCB_results.json',
                'JSON/static_ImpCPD_results.json',
                'JSON/static_M_UCB_results.json',
                 'JSON/static_RExp3_results.json',
                 'JSON/static_Sliding_Window_Thompson_sampling_results.json',
                 'JSON/static_Sliding_Window_UCB_results.json',
                 'JSON/static_UCBL_CPD_results.json']




# Labels for the plot
labels = ['ADR-KLUCB','ADR-TS', 'ADS-TS', 'Discounted UCB',  'ImpCPD', 'M-UCB', 'R-Exp3', 'SW-TS', 'SW-UCB', 'UCBL-CPD']

# --- NEW ---
# Define a list of markers to cycle through, like in the example plot
markers = ['o', 'o', 'o', 'v', '^', '<', '>', 's', 'p', '*', 'D', 'X']

plt.figure(figsize=(10,6))

for i, path in enumerate(static_paths):
    try:
        with open(path, 'r') as f:
            data = json.load(f)

        time_horizon = data['time_horizons'] 
        avg_cum_data = data['average_cumulative_regret'] 
        plt.plot(time_horizon, 
                 avg_cum_data, 
                 label=labels[i], 
                 marker=markers[i % len(markers)], 
                 linestyle='-')

    except FileNotFoundError:
        print(f"Warning: File not found at {path}. Skipping.")
    except KeyError as e:
        print(f"Warning: Key error {e} in file {path}. Check JSON structure. Skipping.")
    except Exception as e:
        print(f"An error occurred with {path}: {e}. Skipping.")


# Check if any data was successfully plotted before saving/showing
if plt.gca().lines:
    plt.xlabel('Time Horizon')
    plt.ylabel('Average Cumulative Regret') 
    plt.title('Static Bandit Algorithms Performance Comparison')
    plt.legend()
    plt.grid(True)
    plt.show()
else:
    print("No data was plotted. Please check your JSON file paths and keys.")
```
